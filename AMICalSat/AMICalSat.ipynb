# AMICal Sat 

A continuación vamos a estudiar el funcionamiento de un cubesat 2U (227*100*100mm) **AMICal Sat** creado por  ["space centre of the Grenoble University"](http://amicalsat.univ-grenoble-alpes.fr/) y ["skobeltsyn institute of nuclear Physics in the lomonosov Moscow state University"](http://www.sinp.msu.ru/en)

<img src='mission.jpg'>

El objetivo general ([misión](http://amicalsat.univ-grenoble-alpes.fr/pages/the_mission)) de este satelite es poder fotografiar las auroras de una orbita LO (low orbit). Los buses de este satelite fueron desarrollados por  SatRevolution. AMICal Sat tiene un transmisor UHF para la banda de satélite Amateur de 70 cm y un transmisor de banda S para la banda de satélite Amateur de 2,4 GHz. El transmisor UHF usa 9k6 FSK y puede transmitir imágenes usando un protocolo [similar](https://destevez.net/2019/08/decoding-swiatowid/) al usado por [Światowid](https://satrevolution.com/missions/swiatowid/). El transmisor de banda S es un transmisor de alta velocidad, por lo que es conveniente utilizarlo como transmisor principal para las imágenes científicas de alta resolución tomadas por la carga útil (Payload), dejando el transmisor UHF como respaldo.

Esta información la he tomado de la pagina [SatRevolution](https://satrevolution.com/missions/amicalsat/) donde podemos encontrar algunas de las características fisicas de el satelite. 

### Payload

- Lentes: 
  - Custome made (hechos a la medida)
  
- sensores: [e2v Onyx](https://imaging.teledyne-e2v.com/products/2d-cmos-image-sensors/onyx/)
  - 1.3 Mpx
  - RGB sparse color matrix
  - AoV 40°
  - Focal length: 23 mm
  - Aperture: f/1.4
  - 12 bit processing

### On-Board Computer 

- 2 × STM32f7 microcontroladores s @216MHz
- MSP430FR microcontrolador @16 MHz
- 2 MB MRAM, 256 kB FRAM
- Ultra Low-Power Modes

### Subsistemas de comunicación y buses 

- **2 × UHF & S-band**
- UHF Frecuencia de operación 435-438 MHz
- S-band Frecuencia de operación 2400-2450 GHz
- UHF rango de transmisíon de datos 9,6 kb/s (opcional de 64 kb/s)
- S-band rango de transmisíon  0,5 Mb/s
- Range 400-500 km

### Guía y control

- Precisión de orientación: ~2 deg
- Máximo offset aleatorio por debajo de 0,5 deg/s/axis

### Systemas eléctricos de potencia

- Payload potencia de salida: 0,5 W
- Voltage de circuito abierto: 6,3 V
- Eficiencia de paneles solares de: 22%

### Chassis 

- 2U estructura 
- CNC manufacturado de Aluminio 7075 alloy
- Anodizado duro y oxidado

### Paquetes de Banda-S 

El transmisor de banda S se basa en un chip transceptor FSK nRF24L01 + 2.4GHz de Nordic Semiconductor (consulte la hoja de datos [aquí](https://www.sparkfun.com/datasheets/Wireless/Nordic/nRF24L01P_Product_Specification_1_0.pdf)). Las imágenes se envían en paquetes Shockburst GFSK de 1 Mbaud. La estructura de un paquete Shockburst se puede ver en la siguiente figura, tomada de la hoja de datos nRF24L01 +.

<img src='paquete.png'>

ALgo importante a tener en cuenta para comunicarnos con el satelite usando la Banda S es que las frecuencias de descarga funcionan de forma constante desde Rusia debido a la ruta de la Órbita, para poder conectarnos con el desde otro punto de la tierra es necesario contactar con la unviersidad para saber en que momento el satelite pasarápor encima de nuestra localización y poder descargar las imagenes.
